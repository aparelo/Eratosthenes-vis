```python
import matplotlib.pyplot as plt
import numpy as np
import time

%matplotlib notebook
```

```python
def erasthotenes(n):
    prime = [True for i in range(n+1)]
    p = 2
    while(p*p <= n):
        if(prime[p] == True):
            for i in range(p * 2, n + 1, p): 
                prime[i] = False
        p += 1
        prime[0] = False
        prime[1] = False
    
    return prime

def sundaram(n):
    prime = [False for i in range(n)]
    marks = [True for i in range(n)]
    
    n_2 = int((n-2)/2)
    
    for i in range(1, n_2+1):
        j = 1
        while((i + j + 2 * i * j) <= n_2): 
            marks[i + j + 2 * i * j] = False
            j += 1
    
    for p, mark in enumerate(marks):
        if(mark and 2*p+1 <= n-1):
            prime[2*p+1] = True
    
    return prime
```

```python
eras_times = []
sun_times = []
n_s = []

for n in range(10000,1000000,10000):
    n_s.append(n)
    start = time.time()
    test = erasthotenes(n)
    runtime = time.time() - start
    eras_times.append(runtime)

    start = time.time()
    test2 = sundaram(n)
    runtime2 = time.time() - start
    sun_times.append(runtime2)


plt.plot(n_s, eras_times, label="erasthotenes")
plt.plot(n_s, sun_times, label="Sundaram")

plt.title("Comparison of sieve algorithm run-times")
plt.ylabel("Run time (s)")
plt.xlabel("Pool size (n)")
plt.legend()
```
